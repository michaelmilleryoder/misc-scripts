# Sample headlines from NOW corpus

```python
# Load sources
# Create metadata df
import csv
import os
from datetime import datetime
import pandas as pd
from tqdm.notebook import tqdm

source_dirpath = '/storage2/mamille3/data/news/now2010-2021/sources/'
all_sources = []
colnames = ['article_id', 'word_count', 'date', 'country', 'source', 'url', 'title']
for fname in tqdm(sorted(os.listdir(source_dirpath))):
    fpath = os.path.join(source_dirpath, fname)
    s = pd.read_csv(fpath, sep='\t', header=None, names=colnames, 
                        parse_dates=['date'], date_parser=lambda x: pd.to_datetime(x, yearfirst=True),
                          index_col='article_id', engine='c', on_bad_lines='skip', quoting=csv.QUOTE_NONE,
                    encoding='latin-1'
                   )
    sampled_sources = s.sample(frac=.05) # 1M articles at 5%
    all_sources.append(sampled_sources)
sources = pd.concat(all_sources).drop_duplicates()
# Filter to 2010-01-01/2020-01-31
sources = sources[sources.date<=datetime(2020,12,31)]
print(sources.shape)
sources.head()
```

```python
# Sample headlines
n = 20000
headlines = sources.title.sample(n=n)
headlines
```

```python
# Save out
headlines.to_csv('/storage2/mamille3/data/news/headlines_20k.txt', index=False, header=False)
```
